```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

üîπ blit=True Âíå blit=False ÁöÑÂå∫Âà´

blit=True	Âè™ÈáçÁªòÊõ¥Êñ∞ÁöÑÈÉ®ÂàÜÔºåÊõ¥È´òÊïàÔºå‰ΩÜ ‰∏çÊîØÊåÅ set_xlim() Âíå set_ylim() ËøôÁßç‰ºöÊîπÂèò Axes ÁªìÊûÑÁöÑÊìç‰Ωú

blit=False	ÊØè‰∏ÄÂ∏ßÈÉΩ‰ºöÈáçÁªòÊï¥‰∏™ÂõæÂÉèÔºåÊÄßËÉΩËæÉ‰ΩéÔºå‰ΩÜÈÄÇÁî®‰∫éÊâÄÊúâ matplotlib ÁâàÊú¨ÔºåÊîØÊåÅ set_xlim() Âíå set_ylim()„ÄÇ

```python
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = ax.plot([], [], 'ro')

def init():
    ax.set_xlim(0, 2*np.pi)
    ax.set_ylim(-1, 1)
    return ln,

def animate(i):
    xdata.append(i)
    ydata.append(np.sin(i))
    ln.set_data(xdata, ydata)   # Ë¶ÜÁõñ‰πãÂâçÁöÑÊï∞ÊçÆ
    return ln,

ani = FuncAnimation(
    fig,
    animate,
    frames=np.linspace(0, 2*np.pi, 128),
    init_func=init,
    blit=True
)
plt.show()
```

# Multi-Line Plot Animation

```python
fig, axes = plt.subplots(1,2)
xdata, ydata1, ydata2 = [], [], []
ln1,  = axes[0].plot([], [], 'ro')
ln2, = axes[1].plot([], [], 'bo')
```

```python
def init():
    axes[0].set_xlim(0, 2*np.pi)
    axes[0].set_ylim(-1, 1)
    axes[1].set_xlim(0, 2*np.pi)
    axes[1].set_ylim(-1, 1)
    return ln1, ln2,

def animate(i):
    xdata.append(i)
    ydata1.append(np.sin(i))
    ydata2.append(np.cos(i))
    ln1.set_data(xdata, ydata1) 
    ln2.set_data(xdata, ydata2)
    return ln1, ln2,

ani = FuncAnimation(
    fig,
    animate,
    frames=np.linspace(0, 2*np.pi, 128),
    init_func=init,
    blit=False
)
plt.show()

```
